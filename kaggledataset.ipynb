```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

drugs = pd.read_csv('/content/drug_side_effects_100k_dataset.csv')

# Convert 'report_date' to datetime objects
drugs['report_date'] = pd.to_datetime(drugs['report_date'])

# Group by date and count the number of side effects reported
daily_reports = drugs.groupby(drugs['report_date'].dt.date).size().reset_index(name='report_count')

plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_reports, x='report_date', y='report_count', color='darkred' ,marker = 'd', markersize = '9' )
plt.title("Trend of Side Effect Reports Over Time")
plt.xlabel("Report Date")
plt.ylabel("Number of Reports")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

gap = px.data.gapminder()

plt.figure(figsize=(12, 6))
sns.lineplot(data=gap, x='year', y='lifeExp', hue='continent')

plt.title('Life Expectancy Over Time by Continent (Gapminder)')
plt.grid(True)
plt.show()

display(gap.head())
```

```python
sns.relplot(data=gap, x='year', y='lifeExp', hue='continent', kind='scatter')
plt.title('Life Expectancy Over Time by Continent (Gapminder)')
plt.grid(True)
plt.show()
```

```python
sns.relplot(data=gap, x='year', y='pop', hue='continent', kind='scatter')
plt.title('Life Expectancy Over Time by Continent (Gapminder)')
plt.grid(True)
plt.show()
```

```python
sns.histplot(data=gap, x='year', hue='continent', multiple='stack')
plt.title('Life Expectancy Over Time by Continent (Gapminder)')
plt.grid(True)
plt.show()
```

```python
sns.barplot(data=gap, x='year', y='lifeExp', hue='continent')
plt.title('Life Expectancy Over Time by Continent (Gapminder)')
plt.show()
```

```python
import matplotlib.pyplot as plt

data = gap[gap['year'] == 2007].groupby('continent')['pop'].sum()

# 2. Plot
plt.figure(figsize=(8, 8))
plt.pie(data, labels=data.index, autopct='%1.1f%%')
plt.title('Population by Continent (2007)')
plt.show()
```
